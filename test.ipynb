```python
page_data = []
font_sizes = []
font_styles = []
```

```python
SAMPLE_PAGES_NUM = 4
```

```python
import random
random_numbers = [random.randint(0, doc.page_count-1) for _ in range(SAMPLE_PAGES_NUM)]
```

```python
random_numbers
```

```python
list(map(lambda x: doc[x], random_numbers))
```

```python
sample_pages = list(map(lambda x: doc[x], random_numbers))
```

```python
for sample_page in sample_pages:
    txt = sample_page.get_text("json")
    raw_data = json.loads(txt)
    for block in raw_data['blocks']:
        try:
            for line in block['lines']:
                for span in line['spans']:
                    print(f"Font size : {span['size']} | Style : {span['font']} | Text : {span['text']}")
                    page_data.append({
                        'font_size' : span['size'],
                        'font_style' : span['font'],
                        'text' : span['text']
                    })
                    font_sizes.append(span['size'])
                    font_styles.append(span['font'])
        except KeyError :
            continue
```

```python
font_para_size = max(set(font_sizes), key=font_sizes.count)
font_para_size
```

```python
font_para_style = max(set(font_styles), key=font_styles.count)
font_para_style
```

```python
prev_type = "subheading"
current_text = ""
```

```python
structured_data = []
```

```python
data = json.loads(doc[6].get_text("json"))
for block in data['blocks']:
    try:
        for line in block['lines']:
            for span in line['spans']: 
                    print(span)
    except KeyError :
        continue
```

```python
for page in doc:
    data = json.loads(page.get_text("json"))
    for block in data['blocks']:
        try:
            for line in block['lines']:
                for span in line['spans']:

                    # if the line is not a para
                    if span['size'] > font_para_size or (span['font'] != font_para_style and 
                                                         span['size'] == font_para_size):
                        current_type = "subheading"
                    else:
                        current_type = "para"

                    if current_type == prev_type:
                        current_text = " ".join([current_text, span['text']])
                    else:
                        structured_data.append({
                            "type" : prev_type,
                            "text" : current_text
                        })
                        prev_type = current_type
                        current_text = span['text']
        except KeyError :
            continue
        
```
